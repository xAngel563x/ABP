##4º Comienzo de pruebas del sistema de recomendación

###4.1 Primero se preprocesa la columna que mantiene el textosobre el que vamos a hacer el sistema de recomendación y se añade una columna a la base de datos con ese texto tokenizado

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer

import nltk
#nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')

ps = PorterStemmer()

preprocessedText = []

for row in comics.itertuples():


    text = word_tokenize(row[5]) ## indice de la columna que contiene el texto
    ## Remove stop words
    stops = set(stopwords.words("english"))
    text = [ps.stem(w) for w in text if not w in stops and w.isalnum()]
    text = " ".join(text)

    preprocessedText.append(text)

comicsPreprocesado = comics
comicsPreprocesado['TextoPreprocesado'] = preprocessedText

comicsPreprocesado.head()
```

```python
comicsPreprocesado.iloc[0]['Descripcion']
```

```python
comicsPreprocesado.iloc[0]['TextoPreprocesado']
```

### 4.2 Vectorización de el texto preprocesado (bags of words)

Se parte de los datos almacenados de comicsPreprocesado, con el campo TextoPreprocesado que contiene la descripción del cómic tokenizado.
El objetivo es pasar esas descripciones tokenizadas a vectores, aplicando a demás la ponderización TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer

bolsaPalabras = TfidfVectorizer()
bolsaPalabras.fit(comicsPreprocesado['TextoPreprocesado'])
textosBoW= bolsaPalabras.transform(comicsPreprocesado['TextoPreprocesado'])
print("Finished")
```

```python
textosBoW.shape
```

```python
print(textosBoW)
```

```python
bolsaPalabras.get_feature_names_out()[3905]#Metodo que sustituyo a get_feature_names_out() en versiones recientes
```

###4.3 Calculo de las distancias entre los vectores de frecuencias

El objetivo es crear una matriz N x N, donde N es el número de cómics, y el valor matriz[a,b] indique la distancia entre el cómic a y b. Se usará la distancia coseno para calcular las distancias.

```python
from sklearn.metrics import pairwise_distances

matrizDistancias = pairwise_distances(textosBoW,textosBoW ,metric='cosine')
```

```python
print(matrizDistancias.shape)
print(type(matrizDistancias))
```

# Prueba del sistema de recomendación (previamente ejecutar celdas de Sistema de Recomendación de Cómics)

```python
tituloBusq = "Black Panther (2016) #3" #Comic base para las recomendaciones
indiceComic = comicsPreprocesado[comicsPreprocesado['Titulo']==tituloBusq].index.values[0]
indiceComic
```

```python
distance_scores = list(enumerate(matrizDistancias[indiceComic]))#Se crea una lista con las distancias
ordered_scores = sorted(distance_scores, key=lambda x: x[1])#Se ordenan las distancias
top_scores = ordered_scores[1:11]#Se escojen las 10 mejores distancias
top_indexes = [i[0] for i in top_scores]#Se buscan los indices de las peliculas con las 10 mejores distancias
comicsPreprocesado['Titulo'].iloc[top_indexes]
```

#Análisis de sentimientos

## 1.Carga de los datos de entrtenamiento
### Se cargan los datos con los que va se entrenar el sistema de analisis de sentimientos

```python

import pandas as pd

trainingData = pd.read_csv('/content/marvel-train.csv')
trainingData = trainingData.head(1000) #Eliminar la funcion head() si se quiere usar todo el dataset. Para las pruebas usamos únicamente los 1000 primeros tweets
trainingData
```

```python
trainingData['label'].value_counts()
```

### Se procesan las valoraciones para realizar el entrenamiento

```python

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
import os
import nltk
nltk.download('punkt_tab')
nltk.download('punkt')
nltk.download('stopwords')


print(os.path.exists(nltk.data.find('tokenizers/punkt')))

ps = PorterStemmer()

textoProcesado = []

for row in trainingData.itertuples():


    text = word_tokenize(row.body) ## indice de la columna que contiene el texto
    ## Remove stop words
    stops = set(stopwords.words("english"))
    text = [ps.stem(w) for w in text if not w in stops and w.isalnum()]
    text = " ".join(text)

    textoProcesado.append(text)

datosProcesados = trainingData
datosProcesados['processed_text'] = textoProcesado

datosProcesados
```

### Se crea la bolsa de palabras

```python
from sklearn.feature_extraction.text import TfidfVectorizer

bolsaSentimientos = TfidfVectorizer()
bolsaSentimientos.fit(datosProcesados['processed_text'])
textsBoW= bolsaSentimientos.transform(datosProcesados['processed_text'])
print("Finished")
```

```python
textsBoW.shape
```

### Entrenamiento del algoritmo de clasificación, en este caso SVM

```python
from sklearn import svm
svc = svm.SVC(kernel='linear') #Modelo de clasificación

X_train = textsBoW #Documentos
Y_train = trainingData['label'] #Etiquetas de los documentos
svc.fit(X_train, Y_train) #Entrenamiento
```

### Se cargan y procesan los documentos de test

```python
testData = pd.read_csv('/content/marvel-test3.csv')
#testData = testData.head(100)   #Descomentar en caso de ejecutar el test con 100 items
testData.columns = ['Indice','Nombre','AñosActivos','Titulo','FechaPublicacion',
                  'Descripcion','Dibujante','Escritor','DibujantePortada','Imprenta',
                  'Formato','Categoría','Precio','label','text']
testData
```

```python
ps = PorterStemmer()

preprocessedText2 = []

for row in testData.itertuples():


    text = word_tokenize(row.text) ## indice de la columna que contiene el texto
    ## Remove stop words
    stops = set(stopwords.words("english"))
    text = [ps.stem(w) for w in text if not w in stops and w.isalnum()]
    text = " ".join(text)

    preprocessedText2.append(text)

datosTestProcesados = testData
datosTestProcesados['processed_text'] = preprocessedText2

datosTestProcesados
```

```python
testData['label'].value_counts()
```

```python
textsBoWTest= bolsaSentimientos.transform(datosTestProcesados['processed_text'])
print("Finished")
```

```python
textsBoWTest.shape
```

### Se hace la clasificación de los documentos de test

```python
X_test = textsBoWTest #Documentos


predictions = svc.predict(X_test) #Se almacena en el array predictions las predicciones del clasificador
```

### Evaluación de las predicciones

```python
from sklearn.metrics import classification_report

Y_test = testData['label'] #Etiquetas reales de los documentos

print (classification_report(Y_test, predictions))
```

### Se puede valorar de forma manual un item poniendo en el label si el comentario te pareció positivo (1), negativo (-1) o neutral(0). Con la siguiente celda de código se sobrescribirá el label.

```python
valoracion = 1       #La valoracion solo puede ser 1, 0 o -1

match valoracion:
    case 1:
        testData.at[indice,'label'] = 1
    case 0:
        testData.at[indice,'label'] = 0
    case -1:
        testData.at[indice,'label'] = -1
    case _:
        print("Valoracion no valida")

testData.iloc[indice]
```
